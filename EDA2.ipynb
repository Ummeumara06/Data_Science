**When to Use Each Scaling Technique:**

**Standard Scaling (Z-Score Normalization):**

How it works: Transforms the data to have a mean of 0 and a standard deviation of 1.

Preferred Scenario:

Algorithms sensitive to feature magnitude: Machine learning models such as Support Vector Machines (SVM), Logistic Regression, K-Nearest Neighbors (KNN), and Principal Component Analysis (PCA) perform better with standardized features because they are distance-based or work with dot products.
Normally distributed data: If your data follows a Gaussian (normal) distribution, standard scaling ensures that all features contribute equally to the model.
Why: Standard scaling ensures that features have the same unit variance, so no single feature disproportionately affects the model.

**Min-Max Scaling:**
How it works: Scales data to a fixed range, typically [0, 1].

Preferred Scenario:

When the algorithm does not assume normally distributed data: For algorithms like Decision Trees or K-Means Clustering, Min-Max scaling is preferred since these algorithms do not rely on normally distributed data.
When dealing with data that has hard limits: If the features have clear boundaries (e.g., percentages, pixel values in images), Min-Max scaling is appropriate to retain these limits in the transformed space.
Why: Min-Max scaling is useful for models where the range of data matters, such as neural networks, where the input data is often expected to be within a range like [0,1].

**Pros and Cons of One-Hot Encoding and Label Encoding**

**One-Hot Encoding:**
How it works: Creates a binary column for each category in a categorical variable.

Pros:

No ordinal relationship assumed: One-hot encoding is ideal for nominal categorical variables where no order or ranking is implied (e.g., color: red, blue, green).
Widely used in most machine learning models: Works well in algorithms like Logistic Regression and Neural Networks.

Cons:

Curse of dimensionality: For categorical variables with many unique values, one-hot encoding creates a large number of new columns, which can increase memory usage and computational cost.
Sparsity: One-hot encoded features result in many zeros, leading to a sparse matrix.

**Label Encoding:**
How it works: Assigns a unique integer to each category.

Pros:

Efficient in terms of space: Requires only one column to represent categorical data, unlike one-hot encoding.
Better for ordinal categorical variables: Label encoding works well when the categories have an implicit order (e.g., low, medium, high).

Cons:

Imposes an ordinal relationship: This is problematic for nominal categories (e.g., countries), where the algorithm may mistakenly infer a ranking from the numbers assigned to the categories.
Bias: Models can learn patterns based on the numerical order imposed by label encoding, leading to bias in certain models like linear regression or distance-based models.

### Feature Engineering:

```python
import numpy as np

# feature1: Work hours by education: A feature that combines education_num and hours_per_week.
# feature2: Log transformation for skewed columns like capital_gain.

df['work_hrs_by_edu']=df['education_num']*df['hours_per_week']

df['log_transform']=df['capital_gain'].apply(lambda x: np.log(x+1))

df[['work_hrs_by_edu','log_transform']].head()
```

**Justification for Log Transformation**
Log transformation is used to handle skewness in data. For example, in our dataset, capital_gain has values that are highly skewed (i.e., a majority of people have 0 capital gain, but some have extremely high values).

Reasons for Log Transformation:
* Reduces Skewness: Log transformation compresses the range of data, especially for highly skewed features, and brings the distribution closer to a normal distribution.
* Improves Model Accuracy: Many machine learning models (like linear regression) assume normally distributed input data. Log transformation helps in meeting this assumption.
* Prevents Outliers from Dominating: Extremely large values are dampened by the log function, preventing a small number of large values from disproportionately affecting model performance.

### Feature Selection

```python
from sklearn.ensemble import IsolationForest

isolation_forest=IsolationForest(contamination=0.05)
outliers=isolation_forest.fit_predict(df[numerical_columns])

df_cleaned=df[outliers==1]
```

**How Outliers Can Affect Model Performance**

Impact on Different Models:
* Linear Models (e.g., Linear Regression): Outliers can skew the fit of the model. For example, a single outlier with a very large residual error can significantly distort the coefficients, leading to poor predictions.
* Distance-Based Models (e.g., KNN, SVM): Outliers can disproportionately influence distance calculations, leading to misclassifications or incorrect clustering.
* Tree-Based Models (e.g., Decision Trees): These models are less sensitive to outliers since they make splits based on certain thresholds, but outliers can still affect the depth and quality of the splits.

Benefits of Removing Outliers:
1. Improved Accuracy: By removing outliers, the model can focus on the central trend in the data.
2. Reduced Variance: Outliers tend to increase variance, which can lead to overfitting in some models.
3. More Robust Predictions: A model trained on data without extreme values is often more robust and generalizes better to unseen data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import ppscore as pps

# Correlation matrix
corr_matrix = df.corr(numeric_only=True)

# PPS matrix
pps_matrix = pps.matrix(df)[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')

# Plot both matrices side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Correlation heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=axes[0])
axes[0].set_title('Correlation Matrix')

# PPS heatmap
sns.heatmap(pps_matrix, annot=True, cmap='coolwarm', ax=axes[1])
axes[1].set_title('PPS Matrix')

plt.show()
```

**PPS (Predictive Power Score):**
* What it measures: PPS is a data-driven method that calculates how well one feature can predict another, without assuming any specific relationship (linear or otherwise).
* Value range: PPS ranges from 0 (no predictive power) to 1 (perfect predictive power).
* Why itâ€™s useful: PPS can capture non-linear relationships between features that Pearson correlation might miss.

**Correlation Matrix:**
* What it measures: The correlation matrix measures the linear relationship between two variables.
* Value range: Pearson correlation coefficient ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation).
* Limitation: It only captures linear relationships and may miss strong non-linear dependencies.

**Comparison of PPS and Correlation:**

Correlation Matrix: Helps identify linear relationships (e.g., age and hours_per_week may have a moderate positive correlation).

PPS: Detects relationships that are not necessarily linear. For example, features like occupation and income may have a high PPS score even if their correlation is low because the relationship may be non-linear.
