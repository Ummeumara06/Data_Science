## Descriptive Analytics for Numerical Columns

```python
# Descriptive Analytics for Numerical Columns 
# Reading the data in the csv file by importing pandas library
import pandas as pd
df=pd.read_csv(r"C:\Users\khanu\Documents\classexcelr\assignment downloads\Basic stats - 1\sales_data_with_discounts.csv")
df.head()
```

```python
# Identifying the numerical columns in the given dataset 
numerical_columns=df.select_dtypes(include=['float64','int64']).columns
print(numerical_columns)
```

```python
# Describing the data  
df.describe()
```

```python
for col in numerical_columns:
    # finding the mean,median,mode and standard deviation of the each column
    mean_col= df[col].mean()
    median_col=df[col].median()
    mode_col=df[col].mode()[0]
    std_col=df[col].std()
    print(f"Column : {col}")
    print(f"Mean : {mean_col}")
    print(f"Median : {median_col}")
    print(f"Mode : {mode_col}")
    print(f"Standard deviation : {std_col}")
```

## Data Visualization


### 1.Histogram

```python
# importing the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for each numerical columns
for col in numerical_columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df[col],kde=True)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Histogram of {col}")
    plt.show()

    # finding the skewness of each column ( Indicates whether the data is skewed (positively or negatively))
    skewness=df[col].skew()
    print(f"{col} Skewness = {skewness}")

    # If the value is greater than 0 then it is positively skewed, if less than 0 it is negatively skewed else normally distributed
    if skewness>0:
        print(f"{col} is positively skewed")
    elif skewness<0:
        print(f"{col} is negatively skewed")
    else:
        print(f"{col} is normally distributed")
    print("\n")
```

### 2. Box Plot

```python
# Box plot to find outliers
for col in numerical_columns:
    # plotting the boxplot
    plt.figure(figsize=(8,5))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()

    # calculating the first quartile, third quartile and the inter quartile range
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    # calculating the lower bound and upper bound
    lower_bound= Q1-1.5*IQR
    upper_bound= Q3+1.5*IQR
    # finding the outliers (the values less than the lower bound and the values greater than upper bound are the outliers)
    outliers = df[(df[col] <lower_bound)| (df[col]>upper_bound)]
    # displaying the outliers
    print(f"{col} Outliers : \n {outliers[col]}")
    print("\n")
```

### 3.Bar chart

```python
# Identifying the categorical columns in the given dataset 
categorical_columns=df.select_dtypes(include=['object','category']).columns
print(categorical_columns)
```

```python
# plotting the bar chart for all the categorical columns
for col in categorical_columns:
    plt.figure(figsize=(9,5))
    sns.countplot(x=df[col],hue=df[col],palette="YlGnBu",order=df[col].value_counts().index)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title(f"Bar plot of {col}")
    plt.xticks(rotation=90)
    plt.show()
    
```

## Standardization of Numerical Variables

### Concept of Standardization (Z-score Normalization):
Standardization involves rescaling the features of a dataset so that they have the properties of a standard normal distribution
with a mean of 0 and a standard deviation of 1.

Z-score Normalization: This is one of the most common methods to standardize data. For a given data point , the z-score is calculated using the formula:  z=(x-mu)/ sigma


where:
x is the value of the feature,
 is the mean of the feature,
 is the standard deviation of the feature

The result is a new feature z that has a mean of 0 and a standard deviation of 1. Standardization is particularly useful when the data follows a normal distribution (bell curve), but it is robust enough to be used in various conditions.

Steps to Standardize Numerical Columns
1. Calculate the mean () and standard deviation () for each numerical column.
2. Apply the z-score formula to transform each value in the numerical columns.

```python
# importing all the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler

# Load the dataset
df=pd.read_csv(r"C:\Users\khanu\Documents\classexcelr\assignment downloads\Basic stats - 1\sales_data_with_discounts.csv")

# Displaying first few rows of the dataset before standardisation
print("First few rows of the dataset")
print(df.head())

# Finding the numerical columns
numerical_cols=df.select_dtypes(include=['float64','int64']).columns

# Iterating through all the numerical columns to plot histogram
for col in numerical_cols:
    plt.figure(figsize=(8,5))
    plt.hist(df[col],alpha=0.7,bins=20,color='Blue',edgecolor='Black')
    plt.title(f'Before Standardization of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Intialize the standard scaler
scaler=StandardScaler()

# Fitting and transforming the data with the standardized value
df[numerical_cols]=scaler.fit_transform(df[numerical_cols])

# Iterating through the standardized column to plot the histplot
for col in numerical_cols:
    plt.figure(figsize=(8,5))
    plt.hist(df[col],alpha=0.7,bins=20,color='Red',edgecolor='Black')
    plt.title(f'After Standardization of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Displaying the first few rows of the standardized dataset
print("First few rows of the Standardized data:")
print(df.head())
```

### Before and After Comparisons:
1. Before Standardization: The numerical columns might have different scales, ranges, and units. The histograms will likely show a variety of shapes depending on the data distribution.
2. After Standardization: All numerical columns will be rescaled to have a mean of 0 and a standard deviation of 1. The histograms should reflect this uniform distribution.

## Conversion of Categorical Data into Dummy Variables

**Need for Converting Categorical Data into Dummy Variables (One-Hot Encoding):**
Most machine learning algorithms require numerical input. Categorical data (such as 'Red', 'Blue', 'Green') needs to be converted into a numerical format to be used in these models. This is where one-hot encoding comes into play.

One-Hot Encoding is a method of converting categorical data into a numerical form. It works by creating new binary (0 or 1) columns for each unique category in a categorical column.

**Heres why we need it:**

1. Avoid Numerical Misinterpretation: If we directly convert categories to integers (e.g., 'Red' = 1, 'Blue' = 2), ML models might misinterpret this as an ordinal relationship (where one category is greater than another).
2. Facilitate Algorithm Compatibility: Many algorithms, such as linear regression and distance-based models like K-Nearest Neighbors (KNN), require numerical input and do not handle categorical data natively.
3. Ensure Proper Weight Assignment: One-hot encoding allows algorithms to assign weights to each category individually without assuming any implicit order or scale.

```python
# import necessary libraries
import pandas as pd

# load the dataset
df=pd.read_csv(r"C:\Users\khanu\Documents\classexcelr\assignment downloads\Basic stats - 1\sales_data_with_discounts.csv")

# display original dataset
print("Original Dataset")
print(df.head())

# finding/ fetching the categorical columns in the dataset
categorical_columns=df.select_dtypes(include=['object','category']).columns
print(categorical_columns)

# Apply one hot encoding to the categorical columns
df_encoded=pd.get_dummies(df,columns=categorical_columns)

# display the dataset after applying one hot encoding
print("\nTransformed dataset with one hot encoding\n")
print(df_encoded.head())
```
