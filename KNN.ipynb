```python
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns and encode them
label_encoder = LabelEncoder()

# Apply label encoding to each categorical column in the dataset
for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = label_encoder.fit_transform(df[column])
```

```python
# feature extraction
x=df.drop('type',axis=1)
y=df['type']
```

```python
# splitting the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# fitting the model
from sklearn.neighbors import KNeighborsClassifier

knn_clf=KNeighborsClassifier(n_neighbors=5,weights='distance')
knn_clf.fit(x_train,y_train)
```

```python
y_pred=knn_clf.predict(x_test)
```

```python
# Evaluating the model

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred,zero_division=1))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for K and distance metrics
param_grid = {'n_neighbors': [3, 5, 7, 9],
              'metric': ['euclidean', 'manhattan', 'minkowski']}

# Perform grid search
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=4)
grid_search.fit(x_train, y_train)

# Best parameters
print("Best K:", grid_search.best_params_['n_neighbors'])
print("Best metric:", grid_search.best_params_['metric'])

# Using the best parameters
knn_best = KNeighborsClassifier(n_neighbors=grid_search.best_params_['n_neighbors'], metric=grid_search.best_params_['metric'])
knn_best.fit(x_train, y_train)
y_pred_best = knn_best.predict(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_clf=KNeighborsClassifier(n_neighbors=3,weights='distance',metric='manhattan')
knn_clf.fit(x_train,y_train)
```

```python
y_pred=knn_clf.predict(x_test)
# Evaluating the model

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred,zero_division=1))
```

```python
# after hyperparamter tuning we can see that the accuracy has increased from 52% to 76%
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Reducing to 2 principal components for visualization
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(x_train)
X_test_pca = pca.transform(x_test)

# Retraining the KNN classifier with the reduced features
knn_best.fit(X_train_pca, y_train)

# Creating a mesh grid for visualization
x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

# Predicting for every point in the grid
Z = knn_best.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plotting the decision boundaries
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, edgecolor='k', s=20)
plt.title("KNN Decision Boundaries")
plt.show()
```
