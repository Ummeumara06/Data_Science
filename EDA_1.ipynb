```python
# drops rows with missing values
df_cleaned=df.dropna()
```

```python
# check for missing values after handling 
df_cleaned.isnull().sum()
```

```python
# checking for datatypes of each column
df_cleaned.dtypes
```

```python
# check for outliers
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,9))
sns.boxplot(data=df_cleaned.select_dtypes(["float64",'int64']))
plt.title("Box plot for outlier detection")
plt.show()
```

```python
# Detect outliers using IQR method

# iterating through all the columns to find outliers
for col in df_cleaned:
    Q1=df_cleaned[col].quantile(0.25)
    Q3=df_cleaned[col].quantile(0.75)
    
    IQR=Q3-Q1
    
    lower_bound= Q1-1.5*IQR
    upper_bound= Q3+1.5*IQR
    
    # Detect the outliers
    outliers=df_cleaned[(df_cleaned[col]< lower_bound) | (df_cleaned[col]> upper_bound)]
    
    # print the outliers
    print(f'Outliers in {col}: {outliers[col].values}\n')
    
    # Remove outliers from the dataset
    df_cleaned=df_cleaned[(df_cleaned[col]>=lower_bound) & (df_cleaned[col]<= upper_bound)]
```

### Conclusion
1. **Missing Values**: Rows with missing values were removed, ensuring that only complete data records are included. This step was necessary to maintain data integrity and avoid potential biases that could arise from imputation.

2. **Data Type Verification**: Upon inspection, all data types were found to be correct and consistent, requiring no further adjustments.

3. **Outlier Treatment**: Outliers were identified and treated using the Interquartile Range (IQR) method. This approach helped to remove extreme values that could skew the analysis, resulting in a more reliable dataset.
