```python
# Droping the columns that won't be used in the model
df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)
```

```python
# converting the categorical data into numerical using one hot encoding
df=pd.get_dummies(df,columns=['Sex','Embarked'],drop_first=True)
```

```python
df.head()
```

```python
# feature extraction

x= df.drop('Survived',axis=1)
y=df['Survived']
```

```python
# split the model into train and test dataset

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

## XGBM Classifier

```python
from xgboost import XGBClassifier

model_xg= XGBClassifier()
model_xg.fit(x_train,y_train)
```

```python
y_pred=model_xg.predict(x_test)
```

```python
# Evaluating the model

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,roc_auc_score,roc_curve

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
y_prob = model_xg.predict_proba(x_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_prob)
print('ROC AUC Score:', roc_auc)
```

```python
# Visualize the Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Oranges', xticklabels=['Not Survived', 'Survived'], 
            yticklabels=['Not Survived', 'Survived'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Visualize the ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

## LGBM Classifier

```python
import lightgbm as lgb

clf_lgb=lgb.LGBMClassifier()

clf_lgb.fit(x_train,y_train)
```

```python
y_predict=clf_lgb.predict(x_test)
```

```python
y_prob = clf_lgb.predict_proba(x_test)[:, 1]
```

```python
# Evaluating the model

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,roc_auc_score,roc_curve

print('Accuracy:', accuracy_score(y_test, y_predict))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_predict))
print('Classification Report:\n', classification_report(y_test, y_predict))

roc_auc = roc_auc_score(y_test, y_prob)
print('ROC AUC Score:', roc_auc)
```

```python
# Visualize the Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_predict)
plt.figure(figsize=(6,4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='PuBuGn', xticklabels=['Not Survived', 'Survived'], 
            yticklabels=['Not Survived', 'Survived'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Visualize the ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, color='aqua', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='Hotpink', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

**Interpretation of the Results:**

Accuracy: LightGBM has a slightly better accuracy compared to XGBoost. Both models perform well, but LightGBM edges out.

Precision: LightGBM performs better for class 0 (Not Survived), indicating fewer false positives, while both models are quite similar for class 1 (Survived).

Recall: XGBoost has slightly better recall for class 0, while LightGBM performs better for class 1, which is crucial for identifying survived passengers.

F1-Score: Both models perform comparably in terms of F1-score, indicating a good balance between precision and recall.

ROC AUC Score: The ROC AUC score for XGBoost indicates strong performance, and would expect similar results for LightGBM.

**Strengths and Weaknesses:** 

**XGBoost:**

* Strength: Slightly better recall for class 0 (Not Survived) and similar F1-scores.
* Weakness: Lower precision for class 0 compared to LightGBM.

**LightGBM:**

* Strength: Better precision for class 0 and slightly better recall for class 1, leading to higher overall accuracy.
* Weakness: Slightly lower recall for class 0 compared to XGBoost.
