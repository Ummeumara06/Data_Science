```python
# Encoding categorical variables
from sklearn.preprocessing import LabelEncoder

label_encoder=LabelEncoder()
for column in df.columns:
    df[column]=label_encoder.fit_transform(df[column])
```

```python
# Correlation heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(), annot=True, cmap='cubehelix')
plt.show()
```

```python
# feature extraction

x=df.drop('class',axis=1)
y=df['class']
```

```python
from sklearn.preprocessing import RobustScaler

# Initialize the robust scaler
scaler = RobustScaler()

# Fit and transform the features
X_scaled = scaler.fit_transform(x)
```

```python
# splitting the data into train and test data
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.3,random_state=42)
```

```python
# Visualizing class distribution
sns.countplot(x='class', data=df,color='purple')
plt.title('Class Distribution')
plt.show()
```

```python
# fitting the model
from sklearn.svm import SVC

svm_clf=SVC(kernel='linear',class_weight='balanced')
svm_clf.fit(x_train,y_train)
```

```python
y_pred=svm_clf.predict(x_test)
```

```python
# evaluating the model
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report

print('Accuracy: ',accuracy_score(y_test,y_pred))
print('Confusion Matrix \n',confusion_matrix(y_test,y_pred))
print('Classification Report \n', classification_report(y_test,y_pred))
```

```python

from sklearn.decomposition import PCA

# Reduce the dataset to 2 principal components
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(x_train)
X_test_pca = pca.transform(x_test)

# Train SVM with the reduced data
svm_classifier_pca = SVC(kernel='linear')
svm_classifier_pca.fit(X_train_pca, y_train)

# Predictions and visualization
y_pred_pca = svm_classifier_pca.predict(X_test_pca)

plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_pred_pca, cmap='coolwarm', edgecolors='k')
plt.title('SVM Classification Results')
plt.show()
```

```python
# Tuning hyperparameters
svm_classifier_rbf = SVC(kernel='rbf', C=1.0, gamma='scale')  # Radial basis function kernel
svm_classifier_rbf.fit(x_train, y_train)

# Predictions with the tuned SVM
y_pred_rbf = svm_classifier_rbf.predict(x_test)

# Evaluation
print(classification_report(y_test, y_pred_rbf))
```

```python
# we can see that accuracy has increased to 76% after tuning the parameters
```

```python
kernels = ['linear', 'poly', 'rbf']
for kernel in kernels:
    svm_model = SVC(kernel=kernel)
    svm_model.fit(x_train, y_train)
    y_pred = svm_model.predict(x_test)
    print(f"Kernel: {kernel}")
    print(classification_report(y_test, y_pred))
```

```python
# we can see that accuracy for kernel poly is highest that is 80%, with kernel linear being the least among the three with 69% accuracy
```

**Practical Implications of SVM:**

SVM is a robust classification method, especially effective for binary classification and linearly separable data. In real-world tasks, it can be used in areas like spam detection, image classification, and medical diagnosis. However, it may struggle with large datasets or require careful tuning to avoid overfitting.
